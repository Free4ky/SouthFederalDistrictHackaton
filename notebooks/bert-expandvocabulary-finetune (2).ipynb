```python
nlp = spacy.load("ru_core_news_sm", 
exclude=['morphologizer', 'parser', 'ner', 'attribute_ruler', 'lemmatizer'])
# define spaCY tokenizer
def spacy_tokenizer(document, nlp=nlp):
    # tokenize the document with spaCY
    doc = nlp(document)
    # Remove stop words and punctuation symbols
    tokens = [
        token.text for token in doc if (
        token.is_stop == False and \
        token.is_punct == False and \
        token.text.strip() != '' and \
        token.text.find("\n") == -1)]
    return tokens
```

```python
from datasets import ClassLabel
import random
import pandas as pd
from IPython.display import display, HTML

def show_random_elements(dataset, num_examples=10):
    assert num_examples <= len(dataset), "Can't pick more elements than there are in the dataset."
    picks = []
    for _ in range(num_examples):
        pick = random.randint(0, len(dataset)-1)
        while pick in picks:
            pick = random.randint(0, len(dataset)-1)
        picks.append(pick)
    
    df = pd.DataFrame(dataset[picks])
    for column, typ in dataset.features.items():
        if isinstance(typ, ClassLabel):
            df[column] = df[column].transform(lambda i: typ.names[i])
    display(HTML(df.to_html()))
```

```python
block_size = 128
```

```python
def group_texts(examples):
    # Concatenate all texts.
    concatenated_examples = {k: sum(examples[k], []) for k in examples.keys()}
    total_length = len(concatenated_examples[list(examples.keys())[0]])
    # We drop the small remainder, we could add padding if the model supported it instead of this drop, you can
        # customize this part to your needs.
    total_length = (total_length // block_size) * block_size
    # Split by chunks of max_len.
    result = {
        k: [t[i : i + block_size] for i in range(0, total_length, block_size)]
        for k, t in concatenated_examples.items()
    }
    result["labels"] = result["input_ids"].copy()
    return result
```

```python
from transformers import Trainer, TrainingArguments
```

```python
lm_datasets = tokenized_datasets.map(
    group_texts,
    batched=True,
    batch_size=1000,
    num_proc=4,
)
```

### 3. Проверка модели

```python
class Dataset(torch.utils.data.Dataset):

    def __init__(self, values):
        
        self.tokenizer = BertTokenizer.from_pretrained('../input/southfedetaldistrict-all-data/hack2.0/tokenizer2') 
        self.texts = [self.tokenizer(val,
                                padding='max_length', max_length=512, truncation=True,
                                return_tensors="pt") for val in values]

    def __len__(self):
        return len(self.texts)


    def get_batch_texts(self, idx):
        # Fetch a batch of inputs
        return self.texts[idx]

    def __getitem__(self, idx):

        batch_texts = self.get_batch_texts(idx)

        return batch_texts


def splitDf(df):
    np.random.seed(112)
    df_train, df_val, df_test = np.split(df.sample(frac=1, random_state=42),
                                         [int(.8*len(df)), int(.9*len(df))])
    return df_train, df_val, df_test
```

```python
class Preprocessor:
    def __init__(self):
        self.lemmatize = Mystem().lemmatize
        pronouns = "я, мы, ты, вы, он, она, оно, они, себя, мой, твой, ваш, наш, свой, его, ее, их, то, это, " \
                    "тот, этот, такой, таков, столько, весь, всякий, сам, самый, каждый, любой, иной, другой, " \
                    "кто, что, какой, каков, чей, сколько, никто, ничто, некого, нечего, никакой, ничей, " \
                    "нисколько, кто-то, кое-кто, кто-нибудь, кто-либо, что-то, кое-что, что-нибудь, что-либо, " \
                    "какой-то, какой-либо, какой-нибудь, некто, нечто, некоторый, некий, который".split(", ")
        particles = "в, без, до, для, за, через, над, по, из, у, около, под, о, про, на, к, перед, при, с, " \
                    "между, а, абы, аж, ан, благо, буде, будто, вроде, да, дабы, даже, едва, ежели, если, же, " \
                    "затем, зато, и, ибо, или, итак, кабы, как, когда, коли, коль, ли, либо, лишь, нежели, но, " \
                    "пока, покамест, покуда, поскольку, притом, причем, пускай, пусть, раз, разве, ровно, " \
                    "сиречь, словно, так, также, тоже, только, точно, хоть, хотя, чем, чисто, что, чтоб, чтобы, " \
                    "чуть, якобы, да, ни, тоже, также, не, только, но, и, столько, сколько, то, чтобы, не, то, " \
                    "чтобы, а, зато, однако, или, либо, когда, едва, пока, как, только, лишь, потому, что, так, " \
                    "как, ибо, оттого, чтобы, для, того, с, тем, при, ещё, ой".split(", ")
        self.stopwords = np.unique(np.hstack((list(stopwords("ru")), pronouns, particles)))
    
    def clean(self, text, lemmatize=False):
        if type(text) == str:
            text = text.lower()
            text = re.sub(r"[\W\d_ ]+", " ", text)
            text = re.sub(r"\s+", " ", text)
            text = " ".join(list(filter(lambda x: len(x) > 2 and x not in self.stopwords, 
                                        self.lemmatize(text))))
        else:
            text = ''
        return text
    
    def clean_table(self, df, columns=None, lemmatize=False, export_file=None):
        if columns:
            df[columns] = df[columns].applymap(lambda x: self.clean(x, lemmatize))
        else:
            df = df.applymap(lambda x: self.clean(x, lemmatize))
        if export_file:
            df.to_csv(export_file)
        return df
    
    def clean_values(self, values, lemmatize=False):
        return np.array([self.clean(v) for v in values])
```

```python
from transformers import BertTokenizer

class AutoCorrect(Preprocessor):
    def __init__(self, vectorizer='../input/all-lemmas/tfidf768_all.joblib', 
                technologies='../input/sousfederaldistrict/Справочник. Технологии.xlsx', 
                industries='../input/sousfederaldistrict/Справочник. Отрасли и подотрасли.xlsx',
                vectorized_tech=None,
                vectorized_industr=None):
        super().__init__()
        self.device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
        pd.set_option('mode.chained_assignment', None)
        self.industr_df = pd.read_excel(industries, header=1)
        
        self.tech_df = pd.read_excel(technologies, header=1, 
                                    usecols=[' 1 уровень', '2 уровень', '3 уровень (уровень тегирования участников)'])
        #self.vectorize = load(vectorizer).transform
        self.model = torch.load('../input/southfedetaldistrict-all-data/hack2.0/fine-tuned_bert2.h5'
                                ,map_location=self.device)
        self.model.config.output_hidden_states=True
        self.techs = self.tech_df['3 уровень (уровень тегирования участников)'].values
        self.industrs = self.industr_df['Наименование подотрасли'].values
        if vectorized_tech is not None:
            self.tech_vecs = np.load(vectorized_tech) # path to np file
        else:
            clean_tech = self.clean_values(self.techs)
            self.tech_vecs = self.vectorize(clean_tech)
            np.save('./tech_vecs',self.tech_vecs)
        if vectorized_industr is not None:
            self.industr_vecs = np.load(vectorized_industr) # path to np file
        else:
            clean_industrs = self.clean_values(self.industrs)
            self.industr_vecs = self.vectorize(clean_industrs)
            np.save('./industr_vecs',self.industr_vecs)
        
    def vectorize(self,test_data):
        test = Dataset(test_data)

        test_dataloader = torch.utils.data.DataLoader(test, batch_size=2)

        self.model.to(self.device)
        self.model.eval()

        with torch.no_grad():
            embeddings = None

            for test_input in tqdm(test_dataloader):

                mask = test_input['attention_mask'].to(self.device)
                input_id = test_input['input_ids'].squeeze(1).to(self.device)

                output = self.model(input_id, mask)
                if embeddings is None:
                    embeddings = np.mean(output.hidden_states[0].detach().cpu().numpy(),axis=1)
                else:
                    embeddings = np.vstack((embeddings,np.mean(output.hidden_states[0].detach().cpu().numpy(),axis=1)))

        return embeddings
        
    def __call__(self, companies_csv='../input/sousfederaldistrict/1. Компании.xlsx', export_file='pred_companies.xlsx', vectorized_descr=None):
        companies_df = pd.read_excel(companies_csv)
        if vectorized_descr is not None:
            desc_vecs = np.load(vectorized_descr) # path to np file
        else:
            desc_vecs = self.vectorize(companies_df["Описание компании"])
            np.save('./desc_vec',desc_vecs)
        print(desc_vecs.shape)
        companies_df["pred_tech1"] = ""
        companies_df["pred_tech2"] = ""
        companies_df["pred_tech3"] = ""
        companies_df["pred_industr1"] = ""
        companies_df["pred_industr2"] = ""
        for i, (industr, tech) in enumerate(tqdm(companies_df[['Подотрасль', 'Технология (3 уровень)']].values)):
            tech = tech.strip()

            tech_sim = cosine_similarity([desc_vecs[i]], self.tech_vecs)[0]
            pred_techs = self.techs[np.argsort(tech_sim)[::-1]][:10]
            print(desc_vecs.shape, self.tech_vecs.shape)
            print(pred_techs)
            print(tech)
            if tech not in pred_techs or tech not in self.techs:
                tech = pred_techs[0]
            companies_df.loc[i, ['pred_tech1', 'pred_tech2', 'pred_tech3']] = \
            self.tech_df[self.tech_df['3 уровень (уровень тегирования участников)'] == tech].iloc[0].values
            industr = industr.strip()
            industr_sim = cosine_similarity([desc_vecs[i]], self.industr_vecs)[0]
            pred_industr = self.industrs[np.argsort(industr_sim)[::-1]][:10]
#             print(industr, industr in self.industrs)
            if industr not in pred_industr or industr not in self.industrs:
                industr = pred_industr[0]
            print(pred_industr)
            print(industr)
            return
#             print("pred", industr, industr in self.industrs)
            companies_df.loc[i, ['pred_industr1', 'pred_industr2']] = \
            self.industr_df[self.industr_df['Наименование подотрасли'] == industr].iloc[0].values
        companies_df.to_excel(export_file)        
```
